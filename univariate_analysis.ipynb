```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
telco_df1=pd.read_csv("C:/Users/poorn/OneDrive/Documents/data analyst pagra/python/Customer-Churn.csv")
telco_df=telco_df1.drop("customerID",axis =1)
for i, predictor in enumerate(telco_df.drop(columns=['Churn', 'MonthlyCharges', 'TotalCharges'])): 
  plt.figure(i)
  sns.countplot(telco_df, x=predictor, hue='Churn') #Churn is the legend
  # Add labels to the bars
  ax = plt.gca()
  for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
  plt.xticks(rotation=90)
```
