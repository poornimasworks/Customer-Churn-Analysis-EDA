```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
telco_df=pd.read_csv("C:/Users/poorn/OneDrive/Documents/data analyst pagra/python/Customer-Churn.csv")
telco_df['Churn']=telco_df['Churn'].replace({'Yes':1,'No':0})

#Divide the data into two dataframes--->churned/active customers
telco_df_Churned =telco_df.loc[telco_df.Churn==1]
telco_df_Active =telco_df.loc[telco_df.Churn==0]
```

```python
def uniplot(df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"]=20
    plt.rcParams["axes.titlesize"]=22
    plt.rcParams["axes.titlepad"]=30
    temp=pd.Series(data =hue)
    fig,ax=plt.subplots()
    width=len(df[col].unique())+7+4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax=sns.countplot(data=df,x=col,order=df[col].value_counts().index,hue=hue,palette='bright')
    plt.show()
    
```

```python
uniplot(telco_df_Churned,col='Partner',title='Partner Vs Gender for Churned Customers',hue='gender')
```

```python
uniplot(telco_df_Active,col='Partner',title='Partner Vs Gender for Active Customers',hue='gender')
```

```python
uniplot(telco_df_Churned,col='SeniorCitizen',title='Senior Citizen Vs Gender for Churned Customers',hue='gender')
```

```python
uniplot(telco_df_Active,col='SeniorCitizen',title='Senior Citizen Vs Gender for Active  Customers',hue='gender')
```

```python
uniplot(telco_df_Churned,col='PaymentMethod',title='Payment Method Vs Gender for Churned Customers',hue='gender')
```

```python
uniplot(telco_df_Active,col='PaymentMethod',title='Payment Method Vs Gender for Active Customers',hue='gender')
```

```python
uniplot(telco_df_Churned,col='Contract',title='Contract Vs Gender for Churned Customers',hue='gender')
```

```python
uniplot(telco_df_Active,col='Contract',title='Contract Vs Gender for Active Customers',hue='gender')
```
